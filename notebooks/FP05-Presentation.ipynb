### Introduction

This project aims to explore observed gravity changes before and after significant earthquakes, focusing on the 2013 Lushan Earthquake in Sichuan, China and the 2011 Tohoku-Oki Earthquake off the coast of Japan. Potential underlying mechanisms driving these events will be presented, as well as how these methods can be used in the future.

### Motivations and Research Question

Overall we aim to answer the question: How does gravity change before and after a seismic event and what are possible mechanisms to cause them? 

This is relavent and important because it can aid earthquake detection and mitigation. The group wanted to see if a gravity change before and after an earthquake is a predictable pattern, or if it is only observed on certain occasions. We were especially interested in this because it could lead to a way of predicted large scale earthquakes much earlier than scientists currently can.

### Data Collection

Data was collected through two different research papers. One was about the Lushan Earthquake and the other for the Tohoku-Oki Earthquake.

- The **Lushan data** was obtained from a published, open access paper analyzing gravity field changes around the earthquake from [nature.com](https://www.nature.com/articles/s43247-023-00860-z#:~:text=According%20to%20the%20gravity%20changes,km%20and%201%20km%2C%20respectively). The data is availible by downloading a zip file under "Data Availibility". Opening the datasets file, the gravity change data was found under the Figure 4 excel file.
- Gravity Station coordinates are availible in the same downloaded zipfile in raw data and process code/input
- **Topography data** for Sichuan, China was accessed from the [ETOPO Global Relief Model](https://www.ncei.noaa.gov/products/etopo-global-relief-model) from the bedrock elevation [netCDF file](https://www.ngdc.noaa.gov/thredds/catalog/global/ETOPO2022/60s/60s_bed_elev_netcdf/catalog.html?dataset=globalDatasetScan/ETOPO2022/60s/60s_bed_elev_netcdf/ETOPO_2022_v1_60s_N90W180_bed.nc) in order to map the data.
- The **Tohoku-Oki data** was found from an academic paper that set out to analyze observed gravity field changes induced by the 2011 Tohoku-Oki from [nature.com](https://www.nature.com/articles/ncomms13349). This article had a link where all gravity field data has been published from around the world during the time surrounding the earthquake. The dataset used in the project was data recorded by a superconducting gravimeter in the underground Kamioka Observatory in Kamioka, Japan. Latitude and Longitude are included in the file. 


**Challenges**:
- It was difficult to find enough gravity data from before and after two separate earthquakes. This was important for our purposes because we want to be able to show that an observed change in Earth's gravitational field on the surface is an observed pattern when it comes to earthquakes.

### Observed Gravity Changes

Observed gravity data have been visualized in the figures below. The Lushan data has been presented in a longer time period, between 2010 and 2013. The Tohoku-Oki data has been presented at a much smaller time scale, over the course of three days surrounding the earthquake. More specifically, the data is from March 9th 2011 to March 12th 2011.


**Lushan Gravity Change**: 
- Indicated in red, the area of interest of gravity change has been circled in a dotted line. The Lushan Earthquake center is marked as a red X.
- These gravity changes have been calculated and calibrated by the Modified Bayesian gravity adjustment (MBGA) method and the Bayesian apparent density inversion (BADI) method. These methods are described further in the paper.
- The earthquake occured on April 20, 2013. The first four plots consist of data before the earthquake and the last two consist of data from after the earthquake.
- The change of gravity is high before the earthquake and minimal after the earthquake.


**Tohoku-Oki Gravity Change**:
- Here we can see a time series of the vertical acceleration measured by the underground sensor at the Kamioka observatory, both before and after a series of cleaning methods were performed on the data. Outliers were removed using the z-score, the data was decimated to a sampling rate of 0.1Hz, and a rolling mean window of 30s was applied to smooth the residuals.
- Both graphs were taken on the timescale from March 9th, 2011 at midnight, to March 12th, 2011 at midnight. Time is on the x-axis of both plots in Coordinated Universal Time (UTC). 
- The initial data had to be converted from units of gravitational potential, V, to µGal using a Gravity Calibration term given in the orignal dataset
- The change in gravity fluctuates quite a bit and gets very large right before the earthquake. We can see the earthquake occured around March 11th, 2011 at 05:46 UTC, which is where the data makes some consitently large spikes on the graphs. 
- Gravity change is hard to determine exactly as the earthquake happened and after it occured. Most of the data here is not very precise because the after shocks were immense. 

### Implications

Conclusions drawn from the gravity change data are consistant with an increase in gravity before an earthquake and a decrease after.

Another important implication is the mechanism for this substantial gravity change. The referenced reports identify deep mass transfer as the mechanism for gravity changes. A fluid diffusion model can be developed for the Lushan Earthquake, and it can be concluded that the deep mass (fluids) transfer occured in the earthquake area. This presentation focuses mainly on proposed gravity changes, but mechanisms can be further researched in the referenced reports.

### Limitations and Weaknesses

One weakness of our approach is that the two earthquakes we focused on were analyzed on different time scales. The Lushan Earthquake was analyzed throughout years and the Tohoku-Oki Earthquake was analyzed on the seconds and minutes time scale. The Tohoku-Oki data was also fairly corrupted by the ground shaking that occured directly after the earthquake. This made the data during this time have lots of extreme irregularities, and even after cleaning the data it can not be used to make accurate findings. Another weakness is that we only had data from two earthquakes, we could have looked at more seismic events to make strengthen our conclusions. 

### Findings
To summarize our findings, we found that there is a consistant pattern of an increase in gravity before an earthquake and a decrease in gravity after the earthquake. Even on different time scales, this conclusion is consistant for both. The mechanism for this pattern of change is a deep mass transfer in the earthquake source area. In short, during both of these earthquakes, the movements of the tectonic plates moved so much mass that the observed gravity in the area changed. This makes sense because the typical number we use to for the acceleration due to gravity on Earth is 9.81 m/s^2, which is an averaged value. A visualizaiton of this is shown with mass distributions in the figure below from NASA.



Since gravity is constant no where on Earth, it makes sense that a shifting of mass would make the observed gravity value change.

The significance of these findings are that looking at how the gravity is changing in an area that has a lot of tectonic acitivty could be helpful in detecting earthquakes before they happen. In this case, it would be useful to have many gravity stations that constatly collect data. Then, anomalies can be detected that are out of the normal range that gravity changes in this area. In the future, the change in gravity could be connected mathematically to the amount of force that the tectonic plates hit eachother at and how much mass has the potential to move in the area of question. All in all, this could be a way of detecting earthquakes earlier, which has the potential to prevent the loss of more lives.


### Future Research

Since this project only looked at the gravity data surrounding two earthquakes, it would be necessary to conduct further research on other seismic events around the world in order to gain more knowledge on this pattern. It could also be helpful to see if gravity changes also occur during other seismic events, like volcanic eruptions.

Future research can also be done in this area in terms of seismic gravity monitering. This can be applied to early detection of earthquakes with early warning systems. An example of this in application is the GRACE satellites and artificial intelligence that could detect mass transport months before an earthquake.
