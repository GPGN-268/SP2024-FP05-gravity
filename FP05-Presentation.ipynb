### Introduction

This project aims to explore observed gravity changes before and after significant earthquakes, focusing on the 2013 Lushan Earthquake in Sichuan, China and the 2011 Tohoku-Oki Earthquake in Japan. Gravity trends are consistant with a pattern of gravity increase before and decrease after these seismic events. Potential underlying mechanisms driving these events will be presented, as well as how these methods can be used in the future.

### Motivations and Research Question

Overall we aim to answer the question: How does gravity change before and after a seismic event and what are possible mechanisms to cause them?
